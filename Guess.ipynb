```
import random
print('I am thinking of a number between 1 and 20.')
number = random.randint(1,20)


for amountGuesses in range(1,7):
  print('take a guess')
  guess = int(input())
 
 
  if guess > number :
    print('too high')
  elif guess < number :
    print('too low')
  else:
    break


if guess == number : 
  print('good guess. Only took you ' + str(amountGuesses) + ' guesses :)')
else :
  print('nope. hard luck, it was actually ' + str(number))


  
   
```
